```python
def func(x):
    return x**3 - 6*x**2 - 15*x + 10
```

```python
x = np.linspace(-5,10,150)
fig.set_size_inches(2,2)
plt.plot(x,func(x))

plt.xlabel('')

plt.ylabel('')
```

```python
def func_derivative(x):
    return 3*x**2-12*x-15
```

```python
def descent_and_plot(X_start = 10,learning_rate = 0.01,threshold= 0.001):
    
    X_start = X_start
    learning_rate = learning_rate
    threshold = threshold
    X_next  = X_start - learning_rate*func_derivative(X_start)
    fig = plt.figure()
    fig.set_size_inches(10,8)

    plt.plot(x,func(x), alpha = 0.5)

    plt.scatter([X_start,X_next],
               [func(X_start),func(X_next)], color = 'black', s = 80)
    
    plt.plot([X_start,X_next],
               [func(X_start),func(X_next)], color = 'red')

    
    plt.xlabel =(['f']) 
    plt.ylabel=(['x'])

    
    while abs(X_start - X_next) > threshold:
        X_start = X_next
        X_next  = X_start - learning_rate*func_derivative(X_start)

        plt.scatter([X_start,X_next],
               [func(X_start),func(X_next)], color = 'black', s = 80)
    
        plt.plot([X_start,X_next],
                   [func(X_start),func(X_next)], color = 'red')
        print(f'Найденная точка в конце градиентного спуска это X = {X_next}')
    
        

    
```

```python
descent_and_plot()
```
